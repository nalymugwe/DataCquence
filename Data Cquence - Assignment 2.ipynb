```python
import pandas as pd
url = 'https://think.cs.vt.edu/corgis/datasets/json/airlines/airlines.json'
pd.read_json(url).head()
```

```python
import requests
from pandas import json_normalize
datafile = requests.get(url=url)
df = json_normalize(datafile.json())
df.head()
```

```python
df.sort_values('Statistics.Flights.Delayed', ascending=False).head(5)
```

```python
df.sort_values('Statistics.# of Delays.Security', ascending=False).head(1)
```
